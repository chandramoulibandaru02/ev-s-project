```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
evCategory = pd.read_csv(r"C:\Users\HP\Downloads\EV Sales by Category - 2024..csv")
```

```python
evCategory
```

```python
evCategory.drop(3,axis = 0,inplace = True)
```

```python
plt.pie(evCategory['Total No. of Vehicle'], labels = evCategory['Wheeler Type'],autopct='%1.1f%%',shadow=True,
    textprops={'fontsize': 12},
    explode = [0,0.2,0]
)

plt.axis('equal')
plt.title("Distribution of EV Vehicles")
plt.show()
```

```python
plt.bar(
    evCategory['Wheeler Type'],
    evCategory['Total No. of Vehicle'],
    width = 0.6,
    color = ['yellow', 'violet', 'orange'],
    edgecolor='black',
    linewidth=1.2,
    hatch='/',
    alpha=0.8,
    label=evCategory['Wheeler Type']
    )
plt.legend()
plt.xlabel("Categories")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Type Wise Distribution of EV")
plt.show()
```

```python
pmdriveState = pd.read_csv(r"C:\Users\HP\Downloads\EV SALES UNDER PM E-DRIVE SCHEME.csv")
```

```python
pmdriveState
```

```python
pmdriveState.isna().sum()
```

```python
pmdriveState = pmdriveState.fillna({'e-3W': pmdriveState['e-3W'].mean()})
```

```python
pmdriveState.isna().sum()
```

```python
pmdriveState
```

```python
pmdriveState.drop(36,axis = 0,inplace=True)
```

```python
df_sorted = pmdriveState.sort_values('State/UT')

# Prepare data for plotting
states = df_sorted['State/UT']
e2w_counts = df_sorted['e-2W']
e3w_counts = df_sorted['e-3W']

# Multi-line plot for e-2W and e-3W counts by state
plt.figure(figsize=(12, 6))
plt.plot(states, e2w_counts, marker='o', label='e-2W')
plt.plot(states, e3w_counts, marker='s', label='e-3W')
plt.xticks(rotation=90, ha='center')
plt.xlabel('State/UT')
plt.ylabel('Count of Vehicles Sold')
plt.title('e-2W and e-3W Sales Counts by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
allEVState = pd.read_csv(r"C:\Users\HP\Downloads\State Wise EV Data.csv")
```

```python
allEVState
```

```python
allEVState = allEVState.fillna(0)
```

```python
allEVState.drop(32,axis = 0,inplace=True)
```

```python
allEVState.isna().sum()
```

```python
df_sorted = allEVState.sort_values('State Name')

# Prepare data for plotting
states = df_sorted['State Name']
e2w_counts = df_sorted['Two Wheeler']
e3w_counts = df_sorted['Three Wheeler']
four = df_sorted['Four Wheeler']
goods = df_sorted['Goods Vehicles']
public = df_sorted['Public Service Vehicle']
Special = df_sorted['Special Category Vehicles']
ambu = df_sorted['Ambulance/Hearses']
Const = df_sorted['Construction Equipment Vehicle']


# Multi-line plot for e-2W and e-3W counts by state
plt.figure(figsize=(12, 6))
plt.plot(states, e2w_counts, marker='o', label='e-2W')
plt.plot(states, e3w_counts, marker='s', label='e-3W')
plt.plot(states, four, marker='^', label='e-4W')
plt.plot(states, goods, marker='v', label='Goods')
plt.plot(states, public, marker='D', label='Public Service Vehicles')
plt.plot(states, Special, marker='*', label='Special Category Vehicles')
plt.plot(states, ambu, marker='p', label='Ambulance')
plt.plot(states, Const, marker='h', label='Construction Vehicles')
plt.xticks(rotation=90, ha='center')
plt.xlabel('State/UT')
plt.ylabel('Count of Vehicles Sold')
plt.title('e-2W and e-3W Sales Counts by State/UT')
plt.legend()
plt.tight_layout()
plt.show()
```
